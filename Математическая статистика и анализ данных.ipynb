```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from statsmodels.distributions.empirical_distribution import ECDF

# Шаг 1: Создаем выборку данных
A = np.random.normal(loc=mu, scale=sigma, size=2000) 
mu = 1
sigma = 1

# Шаг 2: Вычисляем выборочные характеристики
mean_A = np.mean(A)
variance_A_biased = np.var(A)  # смещенная выборочная дисперсия
variance_A_unbiased = np.var(A, ddof=1)  # несмещенная выборочная дисперсия
std_dev_A = np.std(A)
quartile_A = np.percentile(A, 50)
median_A = np.median(A)
quantile_95_A = np.percentile(A, 95)
quantile_99_A = np.percentile(A, 99)

# Шаг 3: Вычисляем теретические характеристики
theoretical_mean = mu
theoretical_variance = sigma
theoretical_std_dev = np.sqrt(theoretical_variance)
theoretical_quartile = stats.norm.ppf(0.5, loc=theoretical_mean, scale=theoretical_std_dev)
theoretical_median = stats.norm.ppf(0.5, loc=theoretical_mean, scale=theoretical_std_dev)
theoretical_quantile_95 = stats.norm.ppf(0.95, loc=theoretical_mean, scale=theoretical_std_dev)
theoretical_quantile_99 = stats.norm.ppf(0.99, loc=theoretical_mean, scale=theoretical_std_dev)

# Шаг 4: Сравнение
comparison_data = {
    'Выборочная характеристика': ['Среднее', 'Смещенная дисперсия', 'Несмещенная дисперсия',
                                   'Стандартное отклонение', 'Медиана', 'Квартиль', 'Квантиль 95%', 'Квантиль 99%'],
    'Выборочное значение': [mean_A, variance_A_biased, variance_A_unbiased,
                            std_dev_A, median_A, quartile_A, quantile_95_A, quantile_99_A],
    'Теоретическое значение': [theoretical_mean, theoretical_variance, theoretical_variance,
                              theoretical_std_dev, theoretical_median, theoretical_quartile,
                              theoretical_quantile_95, theoretical_quantile_99]
    }
    
comparison_df = pd.DataFrame(comparison_data)
print(comparison_df)

# Шаг 5: Построение графиков (Эмпирическая и теоретическая функции распределения)
norm_rv = stats.norm(loc=mu, scale=sigma)
x = np.linspace(-10, 15, 500)
cdf = norm_rv.cdf(x)
ecdf = ECDF(A)
y = ecdf(x)

plt.plot(x, cdf, label='Теоретическая функция распределения')
plt.plot(x, y, label='Эмпирическая функция распределения')
plt.legend()
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.grid()
plt.show()

# Шаг 6: Гистограмма и функция плотности
sns.histplot(A, bins=30, kde=True, label='Гистограмма')
plt.plot(x, stats.norm.pdf(x, loc=mu, scale=sigma), label='Функция плотности')
plt.legend()
plt.ylabel('Частота')
plt.xlabel('$x$')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, chi2, t

# Задание 1: Генерация выборки и расчет доверительного интервала для a

# Генерация выборки
sample = np.random.normal(loc=a, scale=np.sqrt(sigma_squared), size=sample_size)
a = 3  
sigma_squared = 9 
sample_size = 2000
confidence_level = 0.95

# Расчет среднего значения и стандартного отклонения
X_bar = np.mean(sample)
sigma = np.std(sample)

# Расчет доверительного интервала для a
z_value = norm.ppf(1 - (1 - confidence_level) / 2)
lower_bound = X_bar - z_value * sigma / np.sqrt(sample_size)
upper_bound = X_bar + z_value * sigma / np.sqrt(sample_size)

print(f"Задание 1:")
print(f"Истинное значение a: {a}")
print(f"Центральный доверительный интервал для a: ({lower_bound}, {upper_bound})")
print()

# Построение графика
plt.figure(figsize=(6, 3))

# Предсказанные значения (среднее)
plt.plot([1, sample_size], [X_bar, X_bar], label='Предсказанные значения', color='green')

# Нижняя граница доверительного интервала
lower_bound_parabola = np.polyfit(range(1, sample_size + 1), [lower_bound] * sample_size, 2)
plt.plot(range(1, sample_size + 1), np.polyval(lower_bound_parabola, range(1, sample_size + 1)),
         label='Нижняя граница ДИ', color='red')

# Верхняя граница доверительного интервала
upper_bound_parabola = np.polyfit(range(1, sample_size + 1), [upper_bound] * sample_size, 2)
plt.plot(range(1, sample_size + 1), np.polyval(upper_bound_parabola, range(1, sample_size + 1)),
         label='Верхняя граница ДИ', color='orange')

plt.legend()
plt.xlabel('Номер наблюдения')
plt.ylabel('Значение')
plt.title('Доверительные интервалы для a')

plt.show()

# Задание 2: Расчет доверительного интервала для sigma^2 при известном a
# (Дисперсия - это sigma^2, поэтому используем chi2-распределение)
sample_variance = np.var(sample)

# Расчет доверительного интервала для sigma^2
chi2_lower_bound = chi2.ppf((1 - confidence_level) / 2, df=sample_size)
chi2_upper_bound = chi2.ppf(1 - (1 - confidence_level) / 2, df=sample_size)
lower_bound_sigma_squared = (sample_size - 1) * sample_variance / chi2_upper_bound
upper_bound_sigma_squared = (sample_size - 1) * sample_variance / chi2_lower_bound

print(f"Задание 2:")
print(f"Истинное значение sigma^2: {sigma_squared}")
print(f"Доверительный интервал для sigma^2: ({lower_bound_sigma_squared}, {upper_bound_sigma_squared})")
print()

# Построение графика
plt.figure(figsize=(6, 3))

# Предсказанные значения (среднее)
plt.plot([1, sample_size], [sigma_squared, sigma_squared], label='Предсказанные значения', color='green')

# Нижняя граница доверительного интервала
lower_bound_parabola = np.polyfit(range(1, sample_size + 1), [lower_bound_sigma_squared] * sample_size, 2)
plt.plot(range(1, sample_size + 1), np.polyval(lower_bound_parabola, range(1, sample_size + 1)),
         label='Нижняя граница ДИ', color='red')

# Верхняя граница доверительного интервала
upper_bound_parabola = np.polyfit(range(1, sample_size + 1), [upper_bound_sigma_squared] * sample_size, 2)
plt.plot(range(1, sample_size + 1), np.polyval(upper_bound_parabola, range(1, sample_size + 1)),
         label='Верхняя граница ДИ', color='orange')

plt.legend()
plt.xlabel('Номер наблюдения')
plt.ylabel('Значение')
plt.title('Доверительные интервалы для sigma^2 (известное a)')

plt.show()

# Задание 3: Расчет доверительного интервала для sigma^2 при неизвестном a
# (Используем теперь распределение chi2 с (n-1) степенями свободы)
lower_bound_sigma_squared_unbiased = (sample_size - 1) * sample_variance / chi2.ppf((1 - confidence_level) / 2, df=sample_size - 1)
upper_bound_sigma_squared_unbiased = (sample_size - 1) * sample_variance / chi2.ppf(1 - (1 - confidence_level) / 2, df=sample_size - 1)

print(f"Задание 3:")
print(f"Истинное значение sigma^2: {sigma_squared}")
print(f"Доверительный интервал для sigma^2 (неизвестное a): ({lower_bound_sigma_squared_unbiased}, {upper_bound_sigma_squared_unbiased})")
print()

# Построение графика
plt.figure(figsize=(6, 3))

# Предсказанные значения (среднее)
plt.plot([1, sample_size], [sigma_squared, sigma_squared], label='Предсказанные значения', color='green')

# Нижняя граница доверительного интервала
lower_bound_parabola = np.polyfit(range(1, sample_size + 1), [lower_bound_sigma_squared_unbiased] * sample_size, 2)
plt.plot(range(1, sample_size + 1), np.polyval(lower_bound_parabola, range(1, sample_size + 1)),
         label='Нижняя граница ДИ', color='red')

# Верхняя граница доверительного интервала
upper_bound_parabola = np.polyfit(range(1, sample_size + 1), [upper_bound_sigma_squared_unbiased] * sample_size, 2)
plt.plot(range(1, sample_size + 1), np.polyval(upper_bound_parabola, range(1, sample_size + 1)),
         label='Верхняя граница ДИ', color='orange')

plt.legend()
plt.xlabel('Номер наблюдения')
plt.ylabel('Значение')
plt.title('Доверительные интервалы для sigma^2 (неизвестное a)')

plt.show()

# Задание 4: Расчет доверительного интервала для a при неизвестном sigma^2
# (Используем t-распределение)
t_value = t.ppf(1 - (1 - confidence_level) / 2, df=sample_size - 1)
lower_bound_a = X_bar - t_value * sigma / np.sqrt(sample_size)
upper_bound_a = X_bar + t_value * sigma / np.sqrt(sample_size)

print(f"Задание 4:")
print(f"Истинное значение a: {a}")
print(f"Доверительный интервал для a (неизвестное sigma^2): ({lower_bound_a}, {upper_bound_a})")

# Построение графика
plt.figure(figsize=(6, 3))

# Предсказанные значения (среднее)
plt.plot([1, sample_size], [X_bar, X_bar], label='Предсказанные значения', color='green')

# Нижняя граница доверительного интервала
lower_bound_parabola = np.polyfit(range(1, sample_size + 1), [lower_bound_a] * sample_size, 2)
plt.plot(range(1, sample_size + 1), np.polyval(lower_bound_parabola, range(1, sample_size + 1)),
         label='Нижняя граница ДИ', color='red')

# Верхняя граница доверительного интервала
upper_bound_parabola = np.polyfit(range(1, sample_size + 1), [upper_bound_a] * sample_size, 2)
plt.plot(range(1, sample_size + 1), np.polyval(upper_bound_parabola, range(1, sample_size + 1)),
         label='Верхняя граница ДИ', color='orange')

plt.legend()
plt.xlabel('Номер наблюдения')
plt.ylabel('Значение')
plt.title('Доверительные интервалы для a (неизвестное sigma^2)')

plt.show()
```
